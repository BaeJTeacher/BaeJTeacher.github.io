```python
import folium


#위도 경도에 해당하는 지도 보기

latitude=37.58762741862614
longitude=126.9818845257969

map=folium.Map(location=[latitude,longitude],zoom_start=17)
map

#마커

folium.Marker([latitude,longitude],popup='눈나무집',tooltip='나의 소울 푸드 가게').add_to(map)
map




#써클마커


folium.CircleMarker([latitude,longitude],color='tomato',radius=50,tooltip='눈나무집').add_to(map)
map

#html코드 삽입

folium.Marker([latitude,longitude],popup='<iframe width="560" height="315" src="https://www.youtube.com/embed/rO5wjLOilR4" title="YouTube video player" frameborder="0" allow="accelerometer; autoplay; clipboard-write; encrypted-media; gyroscope; picture-in-picture" allowfullscreen></iframe>집',tooltip='나의 소울 푸드 가게').add_to(map)
map
```
